x_{0,3} = \begin{pmatrix} 0\\1\\0 \end{pmatrix}, \hspace{.5em}

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import exp 
### CODE FOR YOUR PLOT HERE ###

t = np.arange(0,100,0.1)

# Solutions corresponding to inital conditions x01, x02, x03x x04 are given in one array names solns
solns = np.array([[exp(-t),exp(-t),np.zeros(len(t))], \
                  [0.5*exp(-t)+0.5, 0.5*exp(-t)-0.5,np.zeros(len(t))], \
                  [0.5*exp(-t)-0.5, 0.5*exp(-t)+0.5,np.zeros(len(t))], \
                  [np.zeros(len(t)), np.zeros(len(t)), 10**(-6)*exp(2*t)]])

# Plotting the time variation of each component for different initial condition
for comp in range(solns.shape[1]):
    plt.figure(figsize=(16,8))
    plt.title("Component of x: {}".format(comp))
    for in_codn in range(solns.shape[0]):
        plt.plot(t,solns[in_codn,comp], label = 'initial value x_0,{}'.format(in_codn))
        plt.xlabel("Time")
        plt.ylabel("Amplitude in x{} direction".format(comp))
    plt.legend()



# Plotting the variation of all three components of vector x witeaxh other for all initial condition
plt.figure(figsize=(16,8))
ax = plt.axes(projection='3d')
for in_codn in range(solns.shape[0]):
    ax.plot3D(solns[in_codn,0], solns[in_codn,1], solns[in_codn,2],label='initial value x_0,{}'.format(in_codn))

ax.set_xlabel("Component x1")
ax.set_ylabel("Component x2")
ax.set_zlabel("Component x3")
ax.legend(loc="best")
ax.set_title("All three component variation of solution vector X")


### CODE FOR YOUR PLOT HERE ###
```

Plot the vector field of the dynamical system as a quiver plot in the three projection planes defined respectively by

Explain the solutions observed in 1.2 using the quiver plots.

```python
### CODE FOR YOUR PLOTS HERE ###
A = np.array([[-0.5, -0.5, 0.0],[-0.5, -0.5, 0.0],[0.0, 0.0, 2.0]])

steps = np.arange(-4, 4, 1)

X1, X2, X3 = np.meshgrid(steps, steps, steps)
X = np.array([X1.flatten(), X2.flatten(), X3.flatten()])

## Vector field is dX/dt, which is given as dX/dt = A*X
X_grad = np.matmul(A, X)

for in_codn in reversed(range(solns.shape[1])):
    plt.figure(figsize=(16,8))
    plt.quiver(X[in_codn %3],X[(in_codn+1)%3],X_grad[in_codn %3],X_grad[(in_codn+1)%3])
    plt.xlabel("X{}".format(in_codn %3 + 1))
    plt.ylabel("X{}".format((in_codn+1)%3 + 1))
    plt.title('Keeping X{}=0'.format((in_codn+2)%3 +1 ))
    plt.show()

### CODE FOR YOUR PLOTS HERE ###
```

#### 1.5
Now consider a system with constant input, namely $\dot{x}(t) = Ax(t) + s(t)$ where A is as before and $s(t) = (1,2,0)^T$. Compute the solution for the initial condition $(0,0,0)^T$. Plot the solution projected onto the planes defined by

1. $x_3 = 0$,
2. the plane from 1.4.

What do you observe?

```python
### CODE FOR YOUR PLOTS HERE ### 

A = np.array([[-0.5, -0.5, 0.0],[-0.5, -0.5, 0.0],[0.0, 0.0, 2.0]])

steps = np.arange(-4, 4, 1)

X1, X2, X3 = np.meshgrid(steps, steps, steps)
X = np.array([X1.flatten(), X2.flatten(), X3.flatten()])
s = np.array([1,2,0])

## Vector field is dX/dt, which is given as dX/dt = A*X
X_grad_new = np.matmul(A, X)

for i in range(X_grad_new.shape[1]):
    X_grad_new[:,i]+=s.T


# 1.0  => For x3 = 0
plt.figure(figsize=(16,8))
plt.quiver(X[0],X[1],X_grad_new[0],X_grad_new[1])
plt.xlabel("X1")
plt.ylabel("X2")
plt.title('Keeping X3=0')
plt.show()

################################################################
# Now we can project X and Y on e plane using our projection matrix
X_e = e.T @ (proj_mat @ X)
Y_e_new = e.T @ (proj_mat @ X_grad_new)

plt.figure(figsize=(16,8))
plt.quiver(X_e[0],X_e[1],Y_e_new[0],Y_e_new[1])
plt.xlabel("e1")
plt.ylabel("e2")
plt.title("Projection on e plane")
plt.show()

### CODE FOR YOUR PLOTS HERE ###
```
