Consider a linear neural field given by the equation
$$
    \tau\dot{u}(x,t) = -u(x,t) + \int_{-\infty}^\infty w(x-x^\prime) u(x^\prime,t)\text{d}x^\prime + s(x,t),
$$
for $x, t\in\mathbb{R}$.

#### 2.1
Assume that the input signal is constant and given by
$$
s(x) = \frac{1}{2\sqrt{\pi}d}\exp\left(-\frac{x^2}{4d^2}\right).
$$
Assume further that the interaction kernel is given by the Gabor function
$$
w(x) = \frac{1}{\sqrt{\pi}b} a\left( \exp\left( -\frac{x^2}{4b^2} \right) \cos(k_0 x) \right).
$$
<div class="alert alert-block alert-success">

Derive an equation for the stationary solution of the system.
*Hint:* Assume that the system has a stable solution that does not depend on time.
</div> 

A stationary solution implies that the time dependency is removed, and that the partial time derivative $\frac{\partial u(x,t)}{\partial t} = 0$. 

This implies:
\begin{align*}
  0     &= -u(x) + \int_{-\infty}^{\infty} w(x-x')u(x')\text{d}x' + s(x) \\
  u(x)  &=  
    \int_{-\infty}^{\infty} w(x-x')u(x')\text{d}x' + s(x) \\
\end{align*}

Written explicitly:
\begin{equation}
  u(x) = 
    a\frac{1}{\sqrt{\pi}b} 
    \int_{-\infty}^{\infty} 
    \exp\left[ -\frac{(x-x')^2}{4b^2} \right]
    \cos[k_0 (x-x')]
    u(x')\text{d}x' 
    + \frac{1}{2\sqrt{\pi}d}\exp\left(-\frac{x^2}{4d^2}\right)
\end{equation}

#### 2.2
<div class="alert alert-block alert-success">

Solve the resulting equation from 2.1 for the stationary solution $u^\ast (x)$ using the Fourier transform.

*Hint:* The Fourier transform of the Gaussian function $f(x) = \frac{1}{\sqrt{2\pi}\sigma}\exp\left( - \frac{x^2}{2\sigma^2} \right)$ is given by 
$$
\tilde{f}(k) = \mathcal{F}[f](k) = \exp\left( -\frac{\sigma^2 k^2 }{2} \right).
$$
You can use this identity to find the Fourier transforms of both $s$ and $w$. First, think about how you have to define $\sigma$ in $f$ to get $s$ into the Gaussian form. Then for $w$, use the identity $\cos(x) = \frac{1}{2}(e^{ix}+e^{-ix})$. Then you can use the time-shift property of the Fourier transform to get rid of the exponentials. You will be left with another Gauss-like term that you can treat similarly to $s$ to find its transform.

The time-shift property states that $\mathcal{F}[f(x)\exp(i k_0 x)](k) = \tilde{f}(k-k_0)$.

The solution has a closed form in the frequency domain.
</div>


One can see that the term

$$
    \int_{-\infty}^{\infty} w(x-x')u(x')\text{d}x' =: \mathcal{C}
$$

is equivalent to performing a *convolution* of $u(x)$ with the interaction kernel $w(x)$

$$
    w(x)*u(x)
$$

Such that the equation for the stationary solution derived above (sec. 2.1) takes the form

$$
    u(x)  = w(x)*u(x)+ s(x)
$$

Now, according to the convolution theorem

$$
    \mathcal{F}[w(x)*u(x)]=\tilde{w}(k)\cdot\tilde{u}(k)
$$
where
\begin{align*}
    \tilde{w}(k) &= \mathcal{F}[w(x)](k) \\
    \tilde{u}(k) &= \mathcal{F}[u(x)](k) \\
\end{align*}

with $\tilde{g}(k)=\mathcal{F}[g(x)](k)$ denoting the Fourier transform of an arbitrary function $g(x)$.

Therefore, applying the Fourier transform to the equation for the stationary solution will yield

\begin{align*}
    \mathcal{F}[u(x)] &= \mathcal{F}[w(x)*u(x)] + \mathcal{F}[s(x)] \\
    \tilde{u}(k) &= \tilde{w}(k)\tilde{u}(k) + \tilde{s}(k) \iff
    \tilde{u}(k) = \frac{\tilde{s}(k)}{1-\tilde{w}(k)}
\end{align*}

Therefore, to find the closed form solution, we need to determine the explicit forms of $\tilde{s}(k)$ and $\tilde{w}(k)$.

------------------------------------------------------------------------------------------------------------------------------

The input term $s(x)$ can be rewritten as

\begin{align*}
    s(x) = \frac{1}{\sigma_s\sqrt{2\pi}}\exp\left(-\frac{x^2}{2\sigma_s^2}\right) \qquad \text{with } \sigma_s = \sqrt{2}d
\end{align*}

such that its Fourier transform becomes:

\begin{align*}
    \mathcal{F}\left[s(x)\right] = \tilde{s}(k) = \exp\left(-\frac{\sigma_s^2k^2}{2}\right) = \exp\left[-\frac{\left(\sqrt{2}d\right)^2k^2}{2}\right] = e^{-d^2k^2}.
\end{align*}

------------------------------------------------------------------------------------------------------------------------------

The interaction kernel $w(x)$ is written explicitly as 

$$
    w(x) = a\frac{1}{\sqrt{\pi}b}\exp\left(-\frac{x^2}{4d^2}\right)\cos(k_0x)
$$

Because of Euler's identity we know that $\cos(k_0x) = \frac{1}{2}(e^{ik_0x} + e^{-ik_0x})$, such that

\begin{align*}
    w(x) &= a\frac{1}{2\sqrt{\pi}b}\exp\left(-\frac{x^2}{4d^2}\right)(e^{ik_0x} + e^{-ik_0x}) \\
    &= a \underbrace{\left[\frac{1}{\sqrt{2\pi}\sigma_w}\exp\left(-\frac{x^2}{2\sigma_w^2}\right)\right]}_{:= g(x)}
    (e^{ik_0x} + e^{-ik_0x}) \qquad \text{with } \sigma_w = \sqrt{2}b \\
    & = a\left[g(x)e^{ik_0x} + g(x)e^{-ik_0x}\right]
\end{align*}

By the time-shift property of the Fourier transform, we know that $\mathcal{F}[g(x)\exp(i k_0 x)](k) = \tilde{g}(k-k_0)$, therefore:

\begin{align*}
     \mathcal{F}[w(x)](k) = \tilde{w}(k) &= 
     \mathcal{F}\left\{a\left[g(x)e^{ik_0x} + g(x)e^{-ik_0x}\right]\right\} \\
     &= a\left\{ \mathcal{F}\left[g(x)e^{ik_0x}\right] + \mathcal{F}\left[g(x)e^{-ik_0x}\right]\right\}\\
     &= a\left[\tilde{g}(k-k_0) + \tilde{g}(k+k_0)\right]
\end{align*}


Given that $g(x) = \frac{1}{\sqrt{2\pi}\sigma_w}\exp\left(-\frac{x^2}{2\sigma_w^2}\right)$, with $\sigma_w = \sqrt{2}b$, its Fourier transform takes the form:

$$
    \mathcal{F}\left[g(x)\right] = \tilde{g}(k) = \exp\left(-\frac{\sigma_w^2k^2}{2}\right) = \exp\left[-\frac{\left(\sqrt{2}b\right)^2k^2}{2}\right] = \exp\left(-b^2k^2\right).
$$

Therefore, the full-form of the Fourier transform of $w(x)$ is:

$$
    \tilde{w}(k) = a\left[e^{-b^2(k-k_0)^2} + e^{-b^2(k+k_0)^2}\right]
$$
------------------------------------------------------------------------------------------------------------------------------
Finally, putting everything back together into the equation for the stationary solution outlined above, gives:

\begin{align*}
    \tilde{u}(k) &= \frac{\tilde{s}(k)}{1-\tilde{w}(k)} \\ 
                &= \frac{e^{-d^2k^2}}{1-a\left[e^{-b^2(k-k_0)^2} + e^{-b^2(k+k_0)^2}\right]}\\
\end{align*}


What one can see from this equation is that the neural field, for an appropriate set of parameters, can act simply as an integrator of the input (when $d >> b$), by letting the input term dominate. However, when $b >> d$, the Gaussian terms in the denominator dominate, thereby amplyifying the 'preferred frequency' $k_0$, and generating a spatial sinusoidal pattern at that frequency. The precise ratio between $b$ and $d$ will determine the frequency range that allows amplification.

```python
import numpy as np
import matplotlib.pyplot as plt

def analytic_solution(x,a,b,d,k_0):
    k = np.fft.fftshift(np.fft.fftfreq(x.shape[0])*x.shape[0])
    s = np.exp(-d**2*k**2)
    neg = np.exp(-(b**2)*(k-k_0)**2)
    pos = np.exp(-(b**2)*(k+k_0)**2)
    w = a*( neg + pos )
    u_k = s/(1.1-w)
    return k, u_k

def plot_analytic_solution(x,a,b,d,k_0):
    f,p = analytic_solution(x,a,b,d,k_0)
    plt.plot(f,p,label=f'$k_0 =$ {k_0}')
    plt.xlim([-15,15])
    plt.xlabel('frequency')
    plt.ylabel('amplitude')
    plt.legend()

    
A = -10
B = 10
n = 300
x = np.linspace(A,B,n)


plt.figure(figsize=(10,4))
plt.subplot(2,2,1)
plot_analytic_solution(x,a=1,b=0.6,d=2,k_0=4)
plt.title('d >> b')
plt.subplot(2,2,3)
plot_analytic_solution(x,a=1,b=0.6,d=2,k_0=8)
plt.subplot(2,2,2)
plot_analytic_solution(x,a=1,b=4,d=0.2,k_0=4)
plt.title('b >> d')
plt.subplot(2,2,4)
plot_analytic_solution(x,a=1,b=4,d=0.2,k_0=8)
plt.tight_layout()

```

Simulate the equation to find and plot the stationary solution by performing an Euler approximation, and replacing the integral by a Riemann sum s.t. $\int_A^B f(x)\text{d}x \approx \sum_{i=0}^n f(x_i)\Delta x$ with $x_i = A + i\Delta x$ and $\Delta x = \tfrac{B-A}{n}$.

</div>

From above, the differential equation for the specified parameters is:
$$
    \dot{u}(x,t) = \left[ -u(x,t) + a\frac{1}{\sqrt{\pi}b} 
    \int_{A}^{B} 
    \exp\left[ -\frac{(x-x')^2}{4b^2} \right]
    \cos[k_0 (x-x')]
    u(x',t)\text{d}x'
    + \frac{1}{2\sqrt{\pi}d}\exp\left(-\frac{x^2}{4d^2}\right) \right]\bigg/\tau
$$

Which can be approximated using forward Euler and the Riemann sum as


\begin{align*}
    u(x,t+1) = u(x,t) + \Delta t \Bigg[ 
        & -u(x,t) \\
        & + a\frac{1}{\sqrt{\pi}b} 
        \sum_{j=0}^{n} 
        \exp\left\{ -\frac{[x-(A+ j\Delta x)]^2}{4b^2} \right\}
        \cos\{k_0 [x-(A+ j\Delta x)]\}
        u\left[(A+ j\Delta x),t\right]\Delta x \\
        & + \frac{1}{2\sqrt{\pi}d}\exp\left(-\frac{x^2}{4d^2}\right) \Bigg]\bigg/\tau
\end{align*}
    


```python
import numpy as np
import matplotlib.pyplot as plt

### YOUR CODE HERE ###


def solve_neural_field(t_0,t_sim,dt,A,B,n,a,b,d,k_0,tau,sigma=0):
    # Time-vector
    timevect = np.linspace(t_0,t_sim,int(t_sim/dt))
    
    # Space-vector
    spacevect = np.linspace(A,B,n)
    
    # Potential vectors
    u = np.zeros((timevect.shape[0],spacevect.shape[0]))
    
    # Constant input
    s = 1/(2*np.sqrt(np.pi)*d)*np.exp(-spacevect**2/(4*d**2))
    
    # Space delta
    dx = (B-A)/n
    
    # Scaling factor of w(x)
    K = a/(np.sqrt(np.pi)*b)
   
    #  Euler integration 
    for t,time in enumerate(timevect[:-1]):
        if np.mod(t,len(timevect)/5) == 0:
            print(t,'/',len(timevect),end='\r')

        # Vector storing values of convolution for each point in space
        w_x = np.zeros(spacevect.shape)

        # For every point in space, evaluate the convolution 
        for i,x in enumerate(spacevect):
            conv_x = 0    
            for j,x_j in enumerate(spacevect):

                x_prime = A + j*dx
                x_diff = x-x_prime
                argexp = -((x_diff**2)/(4*b**2))
                argcos = k_0*x_diff
                
                conv_x += np.exp(argexp)*np.cos(argcos)*u[t,j]*dx
            w_x[i] = conv_x

        # Add noise
        S = s + + sigma * np.random.randn(spacevect.shape[0])
        
        # Euler step
        euler_step = (-u[t,:] + K*w_x + S)/tau
        u[t+1,:] = u[t,:] + dt*euler_step
    return spacevect, timevect, u

# ---------------------------- STUFF FOR PLOTTING --------------------------------------
def plot_neural_field(x,t,u,a,b,d,k_0):
    
    # Space-time
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.imshow(u.T)
    plt.xlabel('Time')
    plt.ylabel('Position')
    plt.title('Potential')
    tiklabel = np.linspace(-10,10,5)
    plt.yticks(ticks=np.linspace(0,300,5),labels=tiklabel)
    
    # Steady state
    plt.subplot(1,2,2)
    plt.plot(u[-1,:],label=f'$k_0 = ${k_0}')
    plt.xlabel('Position')
    plt.ylabel('Potential')
    plt.title('Steady-state')
    plt.legend()
    plt.legend()
    plt.tight_layout()
    plt.show()

    
### YOUR CODE HERE ###
```

```python
# Euler parameters
t_0 = 0; t_sim = 25; dt = .1

# Integration parameters
A = -10; B = 10; n = 300

# Equation parameters
a = 1; b = 0.6 ; d = 2; k_0 = 4; tau = 10


x, t, u = solve_neural_field(t_0,t_sim,dt,A,B,n,a,b,d,k_0,tau)

plot_neural_field(x,t,u,a,b,d,k_0)
```

As we saw in the analytic solution, with $d << b$, at steady state the spatially distributed potential $u(x)$ of the neural field simply approximates the spatial input $s(x)$ that it is receiving.
